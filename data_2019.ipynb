```python
# set-up imports
import pandas as pd
from pathlib import Path

# load the 2019 files
June_2019_to_load = Path("\\Users\\evapa\\OneDrive\\Desktop\\201906-citibike-tripdata.csv")
#July_2019_to_load = Path("Resources\\201907-citibike-tripdata.csv")

# read the files and store in Pandas Dataframes
June_2019 = pd.read_csv(June_2019_to_load)
#July_2019 = pd.read_csv(July_2019_to_load)

# combine both 2019 files into one dataset
#data_2019 = pd.merge(June_2019, July_2019, how = "outer", on=["bikeid", "bikeid"])
June_2019.head()
```

```python
# assess the make-up of the data
June_2019.info()
```

```python
# check the data distribution for any numeric values
June_2019.describe().round(2)
```

```python
# changing the start and stop time from objects to a timestamp
June_2019['starttime'] = pd.to_datetime(June_2019['starttime'])
June_2019['stoptime'] = pd.to_datetime(June_2019['stoptime'])
```

```python
# decrease the amount of data stored by changing the repeated values to categorical values
columns = ['start station name', 'end station name', 'bikeid', 'usertype']
for x in columns:
    June_2019[x] = June_2019[x].astype('category')
```

```python
June_2019.info()
```

```python
# change the trip duration from seconds to minutes
June_2019['tripminutes'] = June_2019['tripduration'] // 60
```

```python
# delete tripduration column now that we have it in minutes
June_2019.drop('tripduration', inplace=True, axis=1)
June_2019.head()
```

```python
# replace the numbers in the gender column to description words (unkown, male, female)
June_2019["gender"].replace(0, "unknown", inplace = True)
June_2019["gender"].replace(1, "male", inplace = True)
June_2019["gender"].replace(2, "female", inplace = True)
June_2019.head()
```

```python
# save this cleaned dataframe to a csv for use in Tableau
filepath = Path("Tableau_Resources\\June_2019.csv")
#filepath.parent.mkdir(parents=True, exist_ok=True)
June_2019.to_csv(filepath)
```
