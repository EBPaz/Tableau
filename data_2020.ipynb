```python
# set-up imports
import pandas as pd
from pathlib import Path

# load the 2019 files
June_2020_to_load = Path("\\Users\\evapa\\OneDrive\\Desktop\\202006-citibike-tripdata.csv")
#July_2019_to_load = Path("Resources\\201907-citibike-tripdata.csv")

# read the files and store in Pandas Dataframes
June_2020 = pd.read_csv(June_2020_to_load)
#July_2019 = pd.read_csv(July_2019_to_load)

# combine both 2019 files into one dataset
#data_2019 = pd.merge(June_2019, July_2019, how = "outer", on=["bikeid", "bikeid"])
June_2020.head()
```

```python
# assess the make-up of the data
June_2020.info()
```

```python
# check the data distribution for any numeric values
June_2020.describe().round(2)
```

```python
# changing the start and stop time from objects to a timestamp
June_2020['starttime'] = pd.to_datetime(June_2020['starttime'])
June_2020['stoptime'] = pd.to_datetime(June_2020['stoptime'])
```

```python
# decrease the amount of data stored by changing the repeated values to categorical values
columns = ['start station name', 'end station name', 'bikeid', 'usertype']
for x in columns:
    June_2020[x] = June_2020[x].astype('category')
```

```python
# re-check the types of data each column now has
June_2020.info()
```

```python
# change the trip duration from seconds to minutes
June_2020['tripminutes'] = June_2020['tripduration'] // 60
```

```python
# delete tripduration column now that we have it in minutes
June_2020.drop('tripduration', inplace=True, axis=1)
June_2020.head()
```

```python
# change the gender column data type to string for further cleaning
June_2020["gender"] = June_2020["gender"].astype(str)
```

```python
# replace the numbers in the gender column to description words (unkown, male, female)
June_2020["gender"]=June_2020["gender"].str.replace("0", "unknown")
June_2020["gender"]=June_2020["gender"].str.replace("1", "male")
June_2020["gender"]=June_2020["gender"].str.replace("2", "female")
June_2020.head()
```

```python
# save this cleaned dataframe to a csv for use in Tableau
filepath = Path("Tableau_Resources\\June_2020.csv")
#filepath.parent.mkdir(parents=True, exist_ok=True)
June_2020.to_csv(filepath)
```
